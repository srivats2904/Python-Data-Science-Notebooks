# **5.) Random Forest Method**

```python
label=np.array(train['revenue'])
features=X_train #np.array(train.drop(columns=['revenue'],axis=1))
feature_list=list(train.drop(columns=['revenue'],axis=1).columns)
```

```python
feature_train,feature_test,label_train,label_test=train_test_split(features,label,test_size=0.25,random_state=42)
```

```python
rf=RandomForestRegressor(n_estimators=1000,random_state=42)
rf.fit(feature_train,label_train)
```

```python
predictions=rf.predict(feature_test) 
errors=abs(predictions-label_test)
print('Mean Absolute Error:',round(np.mean(errors), 2),'degrees.')
mape=np.mean(100 * (errors/label_test))
accuracy=100-mape
print('Accuracy:',round(accuracy,2),'%.')
```

# a.) Find the importance of each of the features and visualize them

```python
rf.feature_importances_
```

# b.) Build a new model by only taking the most important metrics

```python
rf_most_important = RandomForestRegressor(n_estimators= 1000, random_state=42)
 
important_indices = [feature_list.index('P1'), feature_list.index('P2')]
train_important =feature_train[:, important_indices]
test_important =feature_test[:, important_indices]
```

```python
rf_most_important.fit(train_important,label_train)
predictions=rf_most_important.predict(test_important)
```

```python
errors=abs(predictions-label_test)
print('Mean Absolute Error:',round(np.mean(errors), 2),'degrees.')
mape=np.mean(100 * (errors/label_test))
accuracy=100-mape
print('Accuracy:',round(accuracy,2),'%.')
```
